```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt

print("Libraries imported!")

dataset_path = "Downloads/dataset-resized/dataset-resized"

if os.path.exists(dataset_path):
    print(f"Dataset found at: {dataset_path}")
    categories = os.listdir(dataset_path)
    print(f"Found {len(categories)} categories: {categories}")
else:
    print("ERROR: Dataset path not found. Please check the 'dataset_path' variable.")
```

```python
# --- Parameters (Hum model ko bata rahe hain ki image kaisi hogi) ---
batch_size = 32
img_height = 180
img_width = 180

# --- 1. Training data banana ---
# Hum TensorFlow ko bata rahe hain ki 'dataset_path' se 80% images 
# training ke liye load karo.
train_ds = tf.keras.utils.image_dataset_from_directory(
  dataset_path,
  validation_split=0.2,  # 20% data validation ke liye rakho
  subset="training",     # Yeh wala hissa training hai
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# --- 2. Validation data banana ---
# Ab hum bachi hui 20% images ko validation ke liye load kar rahe hain.
val_ds = tf.keras.utils.image_dataset_from_directory(
  dataset_path,
  validation_split=0.2,
  subset="validation",   # Yeh wala hissa validation hai
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# --- 3. Check karna ki sab sahi hai ---
class_names = train_ds.class_names
print("\nCategories successfully loaded as class names:")
print(class_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")

print("Displaying 9 sample images from the Training Set...")
plt.show()
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

print("Dataset performance optimization applied.")
```

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

num_classes = len(class_names) # Yeh 6 hai (categories)

# --- 1. Data Augmentation Layer ---
# Yeh model ko training ke waqt images ko thoda ghuma kar (rotate)
# ya flip karke dikhayega, taaki model behtar seekh paaye.
data_augmentation = Sequential(
  [
    layers.RandomFlip("horizontal",
                      input_shape=(img_height,
                                  img_width,
                                  3)),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
  ]
)

# --- 2. CNN Model Banana ---
model = Sequential([
  # Pehle, augmentation layer
  data_augmentation,
  
  # Image pixels ko 0-255 se 0-1 range mein laana
  layers.Rescaling(1./255),
  
  # Pehla Convolutional block
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  
  # Dusra Convolutional block
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  
  # Teesra Convolutional block
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  
  # Data ko 1D line mein badalna
  layers.Flatten(),
  
  # Faisla lene wala layer
  layers.Dense(128, activation='relu'),
  
  # Aakhri output layer (6 categories ke liye)
  layers.Dense(num_classes)
])

# --- 3. Model ko "Compile" karna ---
# Hum model ko bata rahe hain ki use seekhna kaise hai
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# --- 4. Model ka structure print karna ---
print("Model created successfully!")
model.summary()
```

```python
epochs = 10  # Hum dataset ko 10 baar train karenge

history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)

print("\n--- Model Training Complete! ---")
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')

print("Displaying model performance graphs...")
plt.show()
```

```python
# Model ko ek file mein save karna
model_filename = 'waste_sorter_model.h5'
model.save(model_filename)

print(f"Model successfully saved as: {model_filename}")
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
import tensorflow as tf # Yeh line add karna zaroori ho sakta hai

# --- 1. Apna save kiya hua model load karna ---
model = load_model('waste_sorter_model.h5')
print("Model loaded successfully!")

# --- INSTRUCTION FUNCTION (Naya hissa) ---
def get_sorting_instruction(predicted_class):
    """
    Yeh function model ke prediction ke aadhaar par
    user ko nirdesh (instruction) deta hai.
    """
    if predicted_class == 'paper':
        return "Nirdesh: Ise 'Paper Recycling' (Neela) bin mein daalein."
    elif predicted_class == 'cardboard':
        return "Nirdesh: Ise 'Cardboard Recycling' (Neela) bin mein daalein."
    elif predicted_class == 'plastic':
        return "Nirdesh: Ise 'Plastic Recycling' (Neela) bin mein daalein."
    elif predicted_class == 'glass':
        return "Nirdesh: Ise 'Glass Recycling' (Peela) bin mein daalein."
    elif predicted_class == 'metal':
        return "Nirdesh: Ise 'Metal Recycling' (Laal) bin mein daalein."
    elif predicted_class == 'trash':
        return "Nirdesh: Yeh recycle nahi ho sakta. Ise 'General Waste' (Kaala) bin mein daalein."
    else:
        return "Nirdesh: Is item ko pehchaan nahi pa raha hoon."

# --- 2. Apni test image ka path (raasta) batana ---
# !!! YAHAN BADLAV KAREIN (Aapka path pehle se sahi hona chahiye) !!!
# (Yahaan wahi path rehne dein jo aapne pichhli baar daala tha)
test_image_path ="Downloads/test bottle_files/ASTON_1000_Sky_Blue_1.jpg"

# --- 3. Image ko load karna aur model ke liye taiyaar karna ---
try:
    # (Humein img_height aur img_width ko dobara define karna hoga)
    img_height = 180
    img_width = 180
    
    img = load_img(test_image_path, target_size=(img_height, img_width))
    img_array = img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) # Create a batch

    # --- 4. Prediction karna ---
    predictions = model.predict(img_array)
    score = tf.nn.softmax(predictions[0])

    # (Humein class_names ko bhi dobara define karna hoga)
    class_names = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

    # --- 5. Result dikhana (Updated) ---
    predicted_class = class_names[np.argmax(score)]
    confidence = 100 * np.max(score)
    instruction = get_sorting_instruction(predicted_class) # Naya function call

    print("\n--- Prediction Result ---")
    print(f"Pehchaan (Identify): Yeh image {predicted_class} hai.")
    print(f"Confidence (Kitna pakka hai): {confidence:.2f}%")
    print(f"\n{instruction}") # <-- Naya instruction yahaan hai!

except FileNotFoundError:
    print(f"ERROR: Image not found at path: {test_image_path}")
    print("Please check the 'test_image_path' variable.")
except Exception as e:
    print(f"Ek error aa gaya hai: {e}")
```

```python
# Yeh MobileNetV2 ke liye zaroori preprocessing function hai
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input

# Hum dataset ko dobara load kar rahe hain, lekin is baar 
# naye preprocessing function ke saath.

# 1. Training data banana
train_ds_v2 = tf.keras.utils.image_dataset_from_directory(
  dataset_path,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# 2. Validation data banana
val_ds_v2 = tf.keras.utils.image_dataset_from_directory(
  dataset_path,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

# 3. Data ko naye function se process karna
train_ds_v2 = train_ds_v2.map(lambda x, y: (preprocess_input(x), y))
val_ds_v2 = val_ds_v2.map(lambda x, y: (preprocess_input(x), y))

# 4. Data ko optimize karna
AUTOTUNE = tf.data.AUTOTUNE
train_ds_v2 = train_ds_v2.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds_v2 = val_ds_v2.cache().prefetch(buffer_size=AUTOTUNE)

print("Data successfully prepared for MobileNetV2.")
```

```python
# 1. Base Model (MobileNetV2) ko load karna
# Hum Google ka pre-trained model load kar rahe hain.
# include_top=False ka matlab hai ki humein aakhri layer nahi chahiye.
base_model = tf.keras.applications.MobileNetV2(input_shape=(img_height, img_width, 3),
                                               include_top=False,
                                               weights='imagenet')

# 2. Base Model ke layers ko "Freeze" karna
# Hum base model ko dobara train nahi karna chahte,
# hum sirf uske seekhe hue gyaan ka istemaal karna chahte hain.
base_model.trainable = False

# 3. Naya Model Banana (Apne layers jodna)
# Hum base_model ke oopar apne naye layers jodenge.
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
prediction_layer = tf.keras.layers.Dense(num_classes) # num_classes = 6

# Sabko ek saath jodna
inputs = tf.keras.Input(shape=(img_height, img_width, 3))
x = base_model(inputs, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x) # Thoda sa dropout, behtar results ke liye
outputs = prediction_layer(x)

new_model = tf.keras.Model(inputs, outputs)

# 4. Naye Model ko Compile karna
new_model.compile(optimizer='adam',
                  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])

# 5. Naye Model ka structure (summary) print karna
print("New Transfer Learning Model created successfully!")
new_model.summary()
```

```python
epochs = 10  # Hum naye model ko bhi 10 epochs train karenge

# Hum naye data (train_ds_v2, val_ds_v2) par 
# naye model (new_model) ko train kar rahe hain.
history_v2 = new_model.fit(
  train_ds_v2,
  validation_data=val_ds_v2,
  epochs=epochs
)

print("\n--- New Model Training Complete! ---")
```

```python
acc = history_v2.history['accuracy']
val_acc = history_v2.history['val_accuracy']

loss = history_v2.history['loss']
val_loss = history_v2.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy (New Model)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss (New Model)')

print("Displaying new model's performance graphs...")
plt.show()
```

```python
# Naye model ko naye naam se save karna
new_model_filename = 'waste_sorter_model_v2.h5'
new_model.save(new_model_filename)

print(f"New model successfully saved as: {new_model_filename}")
```
